<a id="data_fomatting"></a>
### Data Formatting
I am converting the date field to a standard format and then computing the date difference w.r.t current date. The rational for this is, the time the cat visited the vet could have some influence in the predicting life expectancy. But later found it didn't help much in the prediction and hence removed while building the model

```python
newdate = pd.DataFrame(columns=["dt1","dt2"])
# Convert the column 'date of last vet visit' from string to the datetime format
newdate['dt1'] =  pd.to_datetime(data['date of last vet visit'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
newdate['dt2'] =  pd.to_datetime(data['date of last vet visit'], format='%Y-%d-%m %H:%M:%S', errors='coerce')
newdate['dt2'] = np.where(newdate.dt2.isnull(), newdate.dt1, newdate.dt2) 
data['date of last vet visit'] = newdate['dt2']
```

```python
# Finding the difference w.r.t current date 
data['date_diff'] = (pd.datetime.now().date() - data["date of last vet visit"]).astype('timedelta64[D]')
# Remove spaces in column names
data.columns = data.columns.str.replace('\s+', '_')
# Printing the top rows
data.head()
```

<a id="build"></a>
### Model Building
Here I have built three model
1. Random Forest Classification
2. Random Forest Regression
3. SVM Regression
And finally one more prediction based on the mean of the three predicted values

I have considered this problem as both classification and regression because the life expectancy in this case can be considered as a set of finite discrete values and also suitable for a regression as it can fall anywhere between the range.

In general Random Forest Algorithm works best when we have a large dataset but here it seems to improve the accuraccy marginally hence being used.

```python
#feature_list = ['C(breed)', 'hair_length', 'height', 'number_of_vet_visits', 'weight', 'date_diff']
feature_list = ['C(breed)', 'hair_length', 'height', 'number_of_vet_visits', 'weight']
formula_ml = 'age_at_death ~ ' + ' + '.join(feature_list)
print(formula_ml)
```

```python
# Get the required columns
y, x = dmatrices(formula_ml, data=data, return_type='dataframe')

# Split the data into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

# Convert the dependent variable to a 1D numpy array
y_train = np.asarray(y_train).ravel()
y_test = np.asarray(y_test).ravel()

# Instantiate and fit our model - Random Forest
random_forest_model1 = ske.RandomForestClassifier(n_estimators=100).fit(x_train, y_train)
random_forest_model2 =  ske.RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=2, max_features='auto', max_leaf_nodes=None, min_samples_leaf=2, min_samples_split=2, min_weight_fraction_leaf=0.0, n_estimators=10, n_jobs=1, oob_score=False, random_state=0, verbose=0, warm_start=False).fit(x_train, y_train)

# Predicting using svm
svm_model = SVR(kernel='linear').fit(x_train, y_train)

# Prediction on test
y_pred1 = random_forest_model1.predict(x_test)
y_pred2 = random_forest_model2.predict(x_test)
y_pred3 = svm_model.predict(x_test)

#Outliers is replace with the Median value
#y_pred3 = np.where(y_pred3 > 20, np.median(y_pred3) , y_pred3) 

# Bringing in the fourth prediction based on the mean of prediction from the three models
y_mean = (y_pred1+y_pred2+y_pred3)/3
```

<a id="feature_importance"></a>
### Feature Importance
This method is really useful to identify the most important features. But it wasn't of much use in this case. But anyways just wanted to include this piece of code.

```python
# Building a decesion tree and finding the feature importance

model = DecisionTreeRegressor(max_depth=8, min_samples_split=10, min_samples_leaf=3)
model.fit(x_train, y_train)
model.feature_importances_
features = x_train.columns
importances = model.feature_importances_

indices = np.argsort(importances)

plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), features[indices])
plt.xlabel('Relative Importance')
plt.show()
```

<a id="evaluate"></a>
### Model Evaluation

The model performance are evaluated using the cross-validation technique


```python
# Cross Validation - Random Forest Classification
scores1 = cross_val_score(random_forest_model1, x_train, y_train, scoring='neg_mean_absolute_error', cv=2)
print('Random Forest Classification:' , scores1.mean())
```

```python
# Cross Validation - Random Forest Regression
scores2 = cross_val_score(random_forest_model2, x_train, y_train, scoring='neg_mean_absolute_error', cv=5)
print('Random Forest Regression:' , scores2.mean())
```

```python
# Cross Validation - SVM Regression
scores3 = cross_val_score(svm_model, x_train, y_train, scoring='neg_mean_absolute_error', cv=5)
print('SVM Regression:' , scores3.mean())
```

```python
# Checking the prediction
y_p1 = pd.DataFrame(y_pred1)
y_p2 = pd.DataFrame(y_pred2)
y_p3 = pd.DataFrame(y_pred3)
y_p4 = pd.DataFrame(y_mean)
y_t1 = pd.DataFrame(y_test)
tdf = pd.concat([y_p1, y_p2, y_p3, y_p4, y_t1], axis=1)
tdf.columns = ['Predicted_rf_class', 'Predicted_rf_reg', 'Predicted_svm_linear', 'Mean_Prediction', 'Actual']
print(tdf.head(10))
```

<a id="save_model"></a>
### Saving Model

```python
# save the model to disk
filename1 = 'Model/rf_model_class.sav'
filename2 = 'Model/rf_model_reg.sav'
filename3 = 'Model/svm_model_reg.sav'
pickle.dump(random_forest_model1, open(filename1, 'wb'))
pickle.dump(random_forest_model2, open(filename2, 'wb'))
pickle.dump(svm_model, open(filename3, 'wb'))
```

<a id="meassure_acc"></a>
### Meassuring Accuracy
Even though the accuracy of the Random Forest Regression Model is better than the other as well as the Mean prediction. We will still go with the mean prediction as it could help in preventing the over-fitting problem

```python
meassure2 = mean_absolute_error(y_pred1, y_test)
meassure3 = mean_squared_error(y_pred1, y_test)
print('RF Class - mean_absolute_error', meassure2)
print('RF Class - mean_squared_error', meassure3)


meassure5 = mean_absolute_error(y_pred2, y_test)
meassure6 = mean_squared_error(y_pred2, y_test)
print('RF Reg - mean_absolute_error', meassure5)
print('RF Reg - mean_squared_error', meassure6)

meassure8 = mean_absolute_error(y_pred3, y_test)
meassure9 = mean_squared_error(y_pred3, y_test)
print('SVM Reg - mean_absolute_error', meassure8)
print('SVM Reg - mean_squared_error', meassure9)

meassure11 = mean_absolute_error(y_mean, y_test)
meassure12 = mean_squared_error(y_mean, y_test)
print('Mean Prediction - mean_absolute_error', meassure11)
print('Mean Prediction - mean_squared_error', meassure12)
```

<a id="user_input"></a>
### Getting User Input

```python
df = pd.DataFrame(columns=["Intercept","T.Javanese", "T.Maine_Coon", "T.Manx", "T.Rex", "T.Siamese", "hair_length", "height", "number_of_vet_visits", "weight"])
parts = int(input("Enter the number of entries to be made:"))
ip = 1
for _ in range(parts):
    ip1 = input("Is the cat's Breed:\n 1 - Javanese\n 2 - Maine Coon\n 3 - Manx\n 4 - Rex\n 5 - Siamese\n")
    ip11 = np.where(ip1 == '1', 1, 0) 
    ip12 = np.where(ip1 == '2', 1, 0)
    ip13 = np.where(ip1 == '3', 1, 0)
    ip14 = np.where(ip1 == '4', 1, 0)
    ip15 = np.where(ip1 == '5', 1, 0)
    ip2 = input("Enter Hair Length:")
    ip3 = input("Enter Height:")
    ip4 = input("Enter Number of Vet Visits:")
    ip5 = input("Enter Weight:")
    #ip6 = input("Enter days from last Vet Visit:")
    df1 = pd.DataFrame(data=[[ip, ip11, ip12, ip13, ip14, ip15,ip2,ip3, ip4, ip5]],columns=["Intercept","T.Javanese", "T.Maine_Coon", "T.Manx", "T.Rex", "T.Siamese", "hair_length", "height", "number_of_vet_visits", "weight"])
    df = pd.concat([df,df1], axis=0)

df.index = range(len(df.index))
print(df)
```

<a id="pred_user_input"></a>
### Prediction for User Input
The prediction is based on the use of all three model and the mean value is used for prediction.

```python
# load the model from disk
model1 = pickle.load(open(filename1, 'rb'))
model2 = pickle.load(open(filename2, 'rb'))
model3 = pickle.load(open(filename3, 'rb'))
#Making prediction for the user input
test_pred1 = model1.predict(df)
test_pred2 = model2.predict(df)
test_pred3 = model3.predict(df)
test_pred = (test_pred1 + test_pred2 + test_pred3)/3
print('Life expectancy of the cat: ', test_pred)
print('Actual Value:', y_test[0])
```
